```python
# import skimage
from PIL import Image
from skimage import io, filters
from skimage.color import rgb2gray
```

```python
def examineFilter(ar, _filter, _print = True, **kwargs):
    aro = _filter(ar, **kwargs)
    if _print:
        print(f"shape = {aro.shape:}, type = {aro.dtype}, min = {aro.min()}, max = {aro.max()}")
    return aro
```

```python
ar = io.imread('~/Pictures/felix.jpg')
# ar = rgb2gray(ar)
print(ar.shape, ar.dtype)
Image.fromarray(ar)
# Image.fromarray(ar > 0.4)
```

```python
aro = filters.frangi(ar[:,:,0] / 255)
aroMax = aro.max()
Image.fromarray((aro / aroMax * 255).astype('uint8'))
# io.imshow(aro)
```

```python
aro = examineFilter(ar, filters.gaussian, sigma = 5)
Image.fromarray((aro * 255).astype('uint8'))
```

```python
aro = examineFilter(ar, filters.hessian)
Image.fromarray((aro * 255).astype('uint8'))
```

```python
aro = examineFilter(ar[:,:,0], filters.inverse, impulse_response = lambda x, y : 0)
Image.fromarray((aro * 255).astype('uint8'))
```

```python
aro = examineFilter(ar, filters.laplace)
Image.fromarray((aro * 255).astype('uint8'))
```

```python
aro = examineFilter(ar, filters.median)
Image.fromarray(aro.astype('uint8'))
```

```python
aro = examineFilter(ar, filters.meijering)
Image.fromarray((aro * 255).astype('uint8'))
```

```python
aro = examineFilter(ar[:,:,0], filters.prewitt)
Image.fromarray((aro / aro.max() * 255).astype('uint8'))
```

```python
arm = (ar / 32).astype('uint8') * 32
display(Image.fromarray(arm)) 
aro = examineFilter(arm, filters.rank_order, _print = False)
# Image.fromarray((aro / aro.max() * 255).astype('uint8'))
aro[0]
aro[1]
```

```python
aro = examineFilter(ar[:,:,0], filters.roberts)
# Image.fromarray((aro / aro.max() * 255).astype('uint8'))
Image.fromarray((aro > 0.05))
```

```python
aro = examineFilter(ar, filters.sato)
Image.fromarray((aro / aro.max() * 255).astype('uint8'))
# Image.fromarray((aro > 0.05))
```

```python
aro = examineFilter(ar[:,:,0], filters.scharr)
Image.fromarray((aro / aro.max() * 255).astype('uint8'))
# Image.fromarray((aro > 0.05))
```

```python
aro = examineFilter(ar[:,:,0], filters.sobel)
# Image.fromarray((aro / aro.max() * 255).astype('uint8'))
Image.fromarray((aro > 0.05))
```

```python
threshold = examineFilter(ar, filters.threshold_isodata)
Image.fromarray((ar[:,:,0] > threshold))
```

```python
aro = examineFilter(ar, filters.unsharp_mask, radius = 1)
Image.fromarray((aro * 255).astype('uint8'))
```

```python
aro = examineFilter(ar, filters.window)
Image.fromarray((aro * 255).astype('uint8'))
```
