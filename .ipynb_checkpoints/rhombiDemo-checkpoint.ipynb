```python
def intersectionPoints(lineCount = 2):
    return int(lineCount * (lineCount - 1) / 2)


ind_1 = 1
ind_2 = 2
while ind_2 < 120:
    l1 = intersectionPoints(lineCount = ind_1)
    l2 = intersectionPoints(lineCount = ind_2)
    dif = l2 - l1
    if abs(dif - 2020) < 1:
        print(f"{ind_1}, {ind_2}, {dif}")
    if dif > 2020:
        ind_1 += 1
    else:
        ind_2 += 1
    



def sortFibo(inList):
    inds = range(len(inList))
    inds = sorted(inds, key = lambda x: (x * theta) % 1)
    rev = [0] * len(inList)
    for j in range(len(inList)):
        rev[inds[j]] = inList[j]
    return rev
        
#     return [inList[i] for i in inds]
    
sortFibo(range(13))
```

```python
def hueBetween(t, huePair):
    (hueMin, hueMax) = huePair
    if hueMax < hueMin:
        hueMax += 360
    return ((hueMin + t * (hueMax - hueMin)) / 360) % 1
```
