## Funktiot

```python
def z2xy(z):
    "Transformation of point in the Unit Square to the Unit Circle (radius 2)"
    point = np.exp(2j * np.pi * (z - 1))
    return point

display(z2xy(1 + 0j))

def xy2point(xy):
    return [int(width / 2 + radius * xy.real) * scale, int(height / 2 + radius * xy.imag) * scale]

# display(xy2point((0.5, 0.5)))

def z2point(z):
    return xy2point(z2xy(z))

display(z2point(1))

def getFilePath(frameNumber):
    return f'c:/data/art/img{frameNumber:03}.png'


"""
def myPolygon(draw, rhombus):
    draw.polygon([tuple(c) for c in rhombus.corners2], fill = rhombus.colorOut)
    draw.polygon([tuple(c) for c in rhombus.corners3], fill = rhombus.colorIn)

def myRhombi(draw, rhombus):
    myPolygon(draw, rhombus)
"""

## Kuvan piirto

def hueFromz(z):
    return (- z.real * 1 + z.imag * 3) * 360 % 360

def valueFromz(z):
    return min(100, 400 * abs((1 - abs(z.imag)) ** 3))
        
hueFromz(0.1095679012345679-0.027391975308641976j)
```

## Neliöjoukko lähtötasossa

```python
class rhombusRaw():
    def __init__(self, center, axis):
        self.center = center
        self.axis = np.array(axis)
    def getCorners(self, orientation = 1.0):
        return self.center + orientation * np.dot(self.axis, np.array([[1, 0, -1, 0], [0, 1, 0, -1]]))
    
rhombusRaw(0.5 + 0.5j, [0.5 + 0.5j, 0.5 - 0.5j]).getCorners()
rhombusRaw(0.5 + 0.5j, [0.5 + 0.5j, 0.5 - 0.5j]).getCorners(orientation = 1j)
```

```python
class squares():
    def __init__(self, n):
        self.items = []
        for j in range(n):
            for i in range(n):
                center = (i + 0.5 + (j + 0.5) * 1j) / n
                if (i + j) % 2 == 0:
                    axis = np.array([0.5 + 0.5j, 0.5 - 0.5j]) / n
                else:
                    axis = np.array([-0.5 - 0.5j, 0.5 - 0.5j]) / n
                r = rhombusRaw(center, axis)
                self.items.append(r)
    def get(self, offset = 0, orientation = 1, squareOrientation = 1):
        return [(z.getCorners(squareOrientation) - centerUnitSquare) * orientation + centerUnitSquare + offset for z in self.items] 

# s.get(0.1 + 0.2j)
# s.get(orientation = 0.1j)
s.get(squareOrientation = 1)
```

## Vinoneliö

```python
def drawImg(rhombi):
    im = Image.new("RGB", (width, height), "black")
    draw = ImageDraw.Draw(im)
    for r in rhombi:
        r.draw(draw)
    return im

class rhombus():
    def __init__(self, zs, hueOffset = 0, orientation = 1):
        self.z = np.mean(zs)
        self.orientation = orientation
        k = 1.0
        self.corners1 = [z2point(z) for z in zs]
        self.center = z2xy(self.z)
        self.deltas = [self.center + 2j * np.pi * np.exp(2j * np.pi * (self.z - 1)) * (z - self.z) * self.orientation for z in zs]
        self.deltasIn = [self.center + k * 2j * np.pi * np.exp(2j * np.pi * (self.z - 1)) * (z - self.z) * self.orientation for z in zs]
        self.corners2 = np.array([xy2point(d) for d in self.deltas])
        self.corners3 = np.array([xy2point(d) for d in self.deltasIn])
        hue = hueFromz(self.z + hueOffset)
        valueOut = valueFromz(self.z)
        valueIn = int(valueOut * 0.55)
        self.colorIn = ImageColor.getrgb(f"hsv({int(hue)},100%,{int(valueIn)}%)")
        self.colorOut = ImageColor.getrgb(f"hsv({int(hue)},100%,{int(valueOut)}%)")
    def draw(self, draw):
        draw.polygon([tuple(c) for c in self.corners1], fill = self.colorOut)
#         draw.polygon([tuple(c) for c in self.corners3], fill = self.colorIn)
    def draw_1(self, draw):
        draw.polygon([tuple(c) for c in self.corners2], fill = self.colorOut)
        draw.polygon([tuple(c) for c in self.corners3], fill = self.colorIn)
    def draw_2(self, draw):
        draw.polygon([tuple(c) for c in self.corners2], fill = self.colorIn)
        draw.polygon([tuple(c) for c in self.corners2[1:]], fill = self.colorOut)
    def __str__(self):
        return f"center {self.center}, corners = {self.corners}, cornersDev = {self.corners2}"
    
s = squares(40)
rhombi = [rhombus(square) for square in s.get()]    
rhombi
display(drawImg(rhombi))
```

## Generoidaan kuva

```python
def genImg(frameNumber, imLast = None, squareCount = 24, offset = 0, colorOffset = 0, blur = False, orientation = 1, squareOrientation = 1, save = False):
    
    ss = squares(squareCount).get(offset, orientation, squareOrientation)
    rhombi = [rhombus(s, -colorOffset) for s in ss]
    im = drawImg(rhombi)#, imLast)
#     ret = Image.fromarray((np.asarray(im) * 0.9).astype('uint8'))
#     im = im.filter(ImageFilter.GaussianBlur(radius = scale))
    
    if blur:
        im = im.filter(ImageFilter.BoxBlur(radius = scale))
    im = im.resize((width, height))
    
    if save == True:
        im.save(getFilePath(frameNumber))
    return im

im = genImg(0
       ,squareCount = 48
#        ,offset = -0.4j - 0.75
#        ,colorOffset = 0.3
#        ,blur = False
#        ,orientation = np.exp(np.pi / 2 * 0 / 2 * 1j - 1.8) 
#        ,squareOrientation = np.exp(np.pi / 2 * 2 / 2 * 1j)
        ,squareOrientation = -1j
        ,save = False)

display(im)
# showImage(getFilePath(0))
```
