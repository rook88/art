```python
def genVoronoiImg(points):
    im = Image.new(size=(width, height), mode="RGB")
    draw = ImageDraw.Draw(im)
    colors = mycolors.genMonoChroma(0.5) 
    vor = Voronoi(points)
    vertices = vor.vertices
    fullRegions = [r for r in vor.regions if not -1 in r and r != [] and insideUnitRegion([vertices[v] for v in r])]

    for ind, testRegion in enumerate(fullRegions):
        (h, s, v) = colors[ind % len(colors)]
        fillColor = hsvText(hue = h, saturation = s, value = v)
        testVertices = [tuple((vertices[v] * width).astype(int)) for v in testRegion]
        draw.polygon(testVertices, fill=fillColor, outline="black")

    for ind, testRegion in enumerate(fullRegions):
        testVertices = [tuple((vertices[v] * width).astype(int)) for v in testRegion]
        for i in range(len(testVertices)):
            j = (i + 1) % len(testVertices)
            draw.line([testVertices[i], testVertices[j]], fill = "black", width = 4)
    return im

# points_1 = 
linSpace = np.linspace(-0.2, 1.2, 15)
points_1 = np.array([x + y * 1j for x in linSpace for y in linSpace])
points_1

points_2 = np.array([np.exp(1j * (a + 0.1)) for a in np.linspace(0, 2 * np.pi, 13, endpoint = False)]) / 18
points_raw = [(p1 + p2) for p1 in points_1 for p2 in points_2]
points = [np.array([p.real, p.imag]) for p in points_raw]

genVoronoiImg(points)
```
