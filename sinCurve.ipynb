```python
import numpy as np
from PIL import Image
```

```python
def sinCurve(margin = 50, unitPixels = 400, cCount = 256 * 3, phase = 0, offset = 0):
    ts = np.linspace(-1, 1, cCount * unitPixels * 2 + 1)[1::2]
    xs = np.sin(np.pi * 5 * ts + phase + offset)
    ys = np.cos(np.pi * 6 * ts + phase * 2)
#     xs = np.sin(np.pi * 2 / (ts - phase))
#     ys = np.cos(np.pi * 2 / (ts + phase))
#     xs = np.sin(phase * np.sign(ts - 0.1) * np.log(abs(ts - 0.1)))
#     ys = np.sin(phase * np.sign(ts + 0.1) * np.log(abs(ts + 0.1)))
#     xs = np.cos(phase * np.sign(ts) * np.log(abs(ts)))
    ar = np.zeros((unitPixels * 2 + margin * 2, unitPixels * 2 + margin * 2), dtype = np.uint8)
    for xr, yr in zip(xs, ys):
        x = int(xr * unitPixels) + unitPixels + 50 
        y = int(yr * unitPixels)+ unitPixels + 50
        for dx in [-1, 0, 1]:
            for dy in [-1, 0 , 1]:
                if ar[y + dy, x + dx] < 255 - 2:
                    ar[y + dy, x + dx] += 2
    return Image.fromarray(ar)
sinCurve(phase = -1, offset = np.pi * np.sqrt(5))
```

```python
frameCount = 100
phases = np.linspace(0, 2 * np.pi / 4, frameCount + 1)[:-1]
len(phases)

sinCurve(phase = phases[-1], offset = np.pi * np.sqrt(5))
```

```python
sinCurve(phase = phases[0], offset = np.pi * np.sqrt(5))
```

```python
for i, phase in enumerate(phases):
    im = sinCurve(phase = phase)
    fileName = f'c:/temp/animation/im_{i:03d}.png'
    print(fileName)
    im.save(fileName)
```
