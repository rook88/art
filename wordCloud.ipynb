````python

file_path = 'C:\\temp\\temp\\zip\\82 Johtoryhmän esityslista\\Johtoryhmä - Steering committee\\2018-06-18\\Emma-Maija johtoryhmä 2018-06-11.pdf'

def text_from_pdf_text(file_path):
    ret = ''
    with open(file_path, 'rb') as f:
        text = ''
        pdf = PdfFileReader(f)
        for page_no in range(pdf.getNumPages()):
            page = pdf.getPage(page_no)
            text_iter = page.extractText().strip()
            ret += '\n' + text_iter.strip() 
    return ret.strip()

text = text_from_pdf_text(file_path)
print(text)
````

````python
file_path = r'C:\temp\temp\zip\82 Johtoryhmän esityslista\Johtoryhmä - Steering committee\2018-06-18\Emma-Maija johtoryhmän kokous 2018-06-18.pdf'

def text_from_pdf_image(file_path):
    pages = convert_from_path(file_path, 500)
    ret = ''
    for page in pages:
        ret += '\n' + pytesseract.image_to_string(page, lang="fin")
    return ret

text = text_from_pdf_image(file_path)
print(text)
````

````python
def text_from_pdf(file_path):
    ret = text_from_pdf_text(file_path)
    if ret != '':
        return ret
    else:
        return text_from_pdf_image(file_path)
        
````

````python
file_path = extension_files['pptx'][12]

def text_from_pptx(file_path):
    ret = ''
    prs = Presentation(file_path)
    for slide in prs.slides:
        for shape in slide.shapes:
            if hasattr(shape, "text"):
                ret += '\n' + shape.text
    return ret

text = text_from_pptx(file_path)
print(text)
````

````python
%%writefile c:/temp/temp/voikko/sanat_module.py
import libvoikko
v = libvoikko.Voikko('fi')
    
def my_guess(sana):
    ret = '?'
    an = v.analyze(sana)
#     print(an)
    if an != []:
        ret = an[0]['BASEFORM']
        return ret
    else:
        su = v.suggest(sana)
        if su != []:
            return su[0]
    return ret
````
